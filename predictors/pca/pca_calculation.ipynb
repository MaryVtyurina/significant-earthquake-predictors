```python
import pandas as pd
from sklearn.decomposition import PCA
```

# Fuction for PCA calculation

```python
def pca(dataframe, eq_id, parameters):
    n_components = 2
    res = pd.DataFrame()
    df = dataframe[dataframe.id == eq_id]
    res = df[['id', 'date', 'nday', 'h', 'm'] + parameters]
    X = df.drop(columns=['id', 'date']).to_numpy()
    pca = PCA(n_components=n_components)
    components = pd.DataFrame(pca.fit_transform(X))
    res.reset_index(drop=True, inplace=True)
    components.reset_index(drop=True, inplace=True)
    components.columns = ['comp_1', 'comp_2']
    return pd.concat([res, components],axis=1)
    
```

## Example

```python
path_in = '../../ionosphere_dataset/NOAA/NOAA_datasets_for_ML/dataset_NN_model/sondes_in_ml.csv'
df_inside = pd.read_csv(path_in, sep=',')
cols_non_nan = ['foF2', 'M(D)', 'MUF(D)', 'fmin', 'foEs', 'fminF', 'fxI', "h'F", "h'F2", "h'Es", 'D', 'foEp', 'peak height of F2 layer',
 'foF2p','id', 'date', 'nday', 'h', 'm']
df_inside = df_inside[cols_non_nan]
```
